```python
import re
import pandas as pd
```

```python
f = open('Whatsapp_chat_with_air_india.txt',"r",encoding="utf-8")
```

```python
data= f.read()
```

```python

pattern = r'\[\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}:\d{2}\]\s'
 
```

```python
messages = re.split(pattern,data)[1:]
messages
```

```python
 dates = re.findall(pattern, data)
```

```python
dates
```

```python
# Create DataFrame
df = pd.DataFrame({'user_message': messages, 'message_date': dates})

# Strip square brackets from 'message_date' column and convert to datetime
try:
    df['message_date'] = pd.to_datetime(df['message_date'].str.strip(), format='[%d/%m/%Y, %H:%M:%S]')
    df.rename(columns={'message_date': 'dates'}, inplace=True)
except ValueError as e:
    print("ValueError:", e)
```

```python
# Convert datetime objects to strings
df['dates'] = df['dates'].astype(str)

# Remove brackets and comma from 'dates' column
df['dates'] = df['dates'].str.replace(r'[\[\],]', '', regex=True)

# Convert the strings back to datetime objects
df['dates'] = pd.to_datetime(df['dates'])

# Print the updated DataFrame
print(df)
```

```python
users = []
messages =[]
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group notification')
        messages.append(entry[0])
        
df['user']= users
df['message'] = messages
df.drop(columns =['user_message'], inplace= True)

df.head()
```

```python
# Convert the 'date' column to datetime
df['dates'] = pd.to_datetime(df['dates'])

# Extract the date in the desired format (yyyy-mm-dd)
df['dates'] = df['dates'].dt.strftime('%Y-%m-%d %H:%M:%S')

df.head()
```

```python
# Convert the 'dates' column to datetime objects
df['dates'] = pd.to_datetime(df['dates'], errors='coerce')

# Drop rows with NaT (Not a Time) values if any
df = df.dropna(subset=['dates'])

# Extract year from the datetime objects
df['year'] = df['dates'].dt.year

df
```

```python
import emoji
import regex as re

def extract_emojis(text):
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
    return ''.join(emoji_pattern.findall(text))

# Example usage:
emojis = []
for message in df['message']:
    emojis.extend(extract_emojis(message))
```
